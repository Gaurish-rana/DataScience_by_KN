Seaborn is a Python data visualization library built on top of Matplotlib that makes it easier to create attractive and informative statistical graphs with less code.

It is mainly used to visualize patterns, trends, and relationships in data using charts like bar plots, box plots, violin plots, heatmaps, and pair plots. Seaborn works seamlessly with Pandas DataFrames and automatically applies clean themes and color palettes, making graphs look more professional compared to basic Matplotlib plots.

```python
import seaborn as sns
```

```python
# Basic Plotting with seaborn
tips =sns.load_dataset('tips')
tips
```

```python
# Create Scatter Plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total Bill vs Tip")
plt.show()
```

```python
# Categorial Plots
# Bar Plot 

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar Plot of Total Bill vs Tip")
plt.show()
```

```python
# Box Plot 

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot of Total Bill vs Tip")
plt.show()
```

```python
# Violin Plot 

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin Plot of Total Bill vs Tip")
plt.show()
```

```python
# Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
# Heatmap
corr=tips[['total_bill','tip','size']].corr()
corr

sns.heatmap(corr,annot=True,cmap='coolwarm')
```
