SQL (Structured Query Language) is a standard programming language used to store, retrieve, update, and manage data in relational databases. It is not a database itself, but a language that allows users to communicate with databases by writing queries such as selecting data, inserting records, updating values, or deleting information. Many popular databases like MySQL, PostgreSQL, Oracle, and SQLite all use SQL to handle data operations.

SQLite, on the other hand, is a lightweight relational database management system that uses SQL as its query language. Unlike other databases, SQLite does not require a separate server and stores the entire database in a single file, making it fast, simple, and easy to use. It is widely used in mobile apps, desktop software, and small to medium projects where simplicity and low setup are important.

```python
import sqlite3
```

```python
#Connect to an SQLite database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
# Create a Table
cursor.execute('''
Create  Table If Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text
    )
''')


# Commit the changes
connection.commit()
```

```python
# Insert the data in sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
values('Krish',30,'Data Analyst')
''')

cursor.execute('''
Insert Into employees(name,age,department)
values('Dev',28,'Data Scientist')
''')

cursor.execute('''
Insert Into employees(name,age,department)
values('Ketan',33,'Frontend Developer')
''')

cursor.execute('''
Insert Into employees(name,age,department)
values('Rahul',31,'Data Engineer')
''')

connection.commit()
```

```python
# Query the data from the table
cursor.execute('Select * from employees')
rows=cursor.fetchall()

# Print the quried rows
for row in rows:
    print(row)
```

```python
# Update the data in the table
cursor.execute('''
UPDATE employees
Set age=35
where name="Krish" 
''')

connection.commit()
```

```python
# Query the data from the table
cursor.execute('Select * from employees')
rows=cursor.fetchall()

# Print the quried rows
for row in rows:
    print(row)
```

```python
# Delete the data from the table
cursor.execute('''
Delete from employees
    where name='Dev'
''')

connection.commit()
```

```python
# Query the data from the table
cursor.execute('Select * from employees')
rows=cursor.fetchall()

# Print the quried rows
for row in rows:
    print(row)
```
