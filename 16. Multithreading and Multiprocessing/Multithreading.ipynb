Multithreading is a programming technique where a single process executes multiple threads concurrently, allowing a program to perform more than one task at the same time. Each thread runs independently but shares the same memory space, which makes communication between tasks faster. Multithreading is mainly used to improve application performance and responsiveness, especially when tasks involve waiting, such as input/output operations like file handling, network requests, or database access. It is also useful in applications that need to handle multiple users or tasks simultaneously, such as web servers and GUI applications, where keeping the interface responsive is important. However, multithreading should be used carefully because shared memory can lead to problems like race conditions and deadlocks, and in some cases, the overhead of managing threads can outweigh the benefits.

```python
import threading
import time


# Single Threading
def print_numbers():
    for i in range(5):
        time.sleep(0.4)
        print(f"Number : {i}")


def print_letter():
    for letter in "abcde":
        time.sleep(0.4)
        print(f"Letter : {letter}")


t=time.time()
print_numbers()
print_letter()
finished=time.time()-t
print(finished)
```

```python
import threading
import time


# Multi Threading
def print_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"Number : {i}")


def print_letter():
    for letter in "abcde":
        time.sleep(2)
        print(f"Letter : {letter}")


# Create 2 threads
t1=threading.Thread(target=print_numbers)
t2=threading.Thread(target=print_letter)

t=time.time()

# Start the threads
t1.start() # begin the thread
t2.start()

# Wait for the thread to complete
t1.join() # wait for the thread to finish
t2.join()

finished=time.time()-t
print(finished)
```
