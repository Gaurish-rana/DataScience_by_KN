# **Steps in Data Preprocessing and Analysis**


## **Data Cleaning**
##### Data cleaning means making the dataset correct and usable. In this step, we handle missing values, remove duplicate records, fix wrong data types (like converting text into numbers or dates), and correct inconsistent values (like “male”, “Male”, “M”). It also includes dealing with outliers and removing unnecessary columns. The main goal is to ensure the data is accurate and free from errors.

## **EDA (Exploratory Data Analysis)**
##### EDA means exploring and understanding the dataset before applying machine learning. In this step, we use statistics and graphs to find patterns, trends, and relationships between variables. We check distributions, detect outliers, find correlations, and understand how different features affect the target output. The goal of EDA is to get insights from data and decide what changes or improvements are needed.

## **Feature Engineering**
##### Feature engineering means improving the dataset by creating new features or modifying existing ones to make the model perform better. It includes encoding categorical data (like converting “red/blue” into numbers), scaling numeric values, selecting important features, creating new columns from existing ones (like making age groups), and applying transformations. The main goal is to make the features more meaningful so the machine learning model can learn more effectively.

```python
## Data Cleaning
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
# Conert Review datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

### 19000K==19M

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy['Size']
```

```python
df_copy.head()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',',"$"]
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
# Handling last updated feature
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

## Explore Data

```python
numeric_feature=[feature for feature in df_copy.columns if df_copy[feature].dtype !='O']
categorial_feature=[feature for feature in df_copy.columns if df_copy[feature].dtype =='O']

# Print Columns
print('We have {} numerical features : {}'.format(len(numeric_feature),numeric_feature))
print('We have {} categorial features : {}'.format(len(categorial_feature),categorial_feature))
```

## **3.2 Feature Information**

##### 1. **App** :- Name of the App
##### 2. **Category** :- Category under which the App falls.
##### 3. **Rating** :- Application’s rating on Play Store
##### 4. **Reviews** :- Number of reviews of the App.
##### 5. **Size** :- Size of the App.
##### 6. **Install** :- Number of installs of the App
##### 7. **Type** :- If the App is free/paid
##### 8. **Price** :- Price of the app (0 if it is Free)
##### 9. **Content Rating** :- Appropriate target audience of the App.
##### 10. **Genres** :- Genre under which the App falls.
##### 11. **Last Updated** :- Date when the App was last updated
##### 12. **Current Ver** :- Current version of the application
##### 13. **Android Ver** :- Minimum Android version required to run the App.

```python
# Proportion of count data on categorial columns
for col in categorial_feature:
    print(df_copy[col].value_counts(normalize=True)*100)
    print('-------------------------------------')
```

```python
# Proportion of count data in numerical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0,len(numeric_feature)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numeric_feature[i]],shade=True,color='r')
    plt.xlabel(numeric_feature[i])
    plt.tight_layout()
```

## Observation
- Rating and Year is left skewedwhile Reviews, Size, Installs and Price are right skewed

```python
# Proportion of count data in categorial columns
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of Categorial Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
