## Data Encoding
##### Data encoding is the process of converting categorical or text data into numerical form so that machine learning algorithms can understand and process it. Since most models work only with numbers, encoding helps transform values like names, categories, or labels into meaningful numeric representations without changing their underlying information. Common encoding methods include label encoding, where each category is assigned a unique number, and one-hot encoding, where categories are represented using binary vectors. Data encoding is an essential step in data preprocessing because it makes raw, non-numeric data usable for training accurate machine learning models.

##### 1. Label Encoding and Ordinal Encoding
##### 2. Nominal or One-Hot Encoding
##### 3. Target Guided Ordinal Encoding

```python
# Difference between ordinal and label encoding
```

### Label Encoding: assigns numbers to categories without any ranking.
#### Example: Red=0, Blue=1, Green=2

### Ordinal Encoding: assigns numbers based on proper order/rank.
#### Example: Small=1, Medium=2, Large=3

## Nominal/OHE Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a simple dataframe
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
# Create an instance of Onehotencoder
encoder=OneHotEncoder() # This creates an object (instance) of OneHotEncoder
```

```python
encoded=encoder.fit_transform(df[['color']]).toarray()
```

### It converts the color column into One-Hot Encoded numeric values.

###### Breakdown (very short):

###### df[['color']] → selects the color column as a 2D DataFrame

###### fit_transform() → learns unique colors and converts them into 0s and 1s

###### .toarray() → changes the result from sparse format to a normal NumPy array

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out()) # Converts the encoded NumPy array into a Pandas DataFrame
```

```python
encoder_df
```

```python
pd.concat([df,encoder_df],axis=1)
```

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
# Create an instance of OrdinalEncoder and then fit_transform
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

## Target Guided Ordinal Encoding

```python
import pandas as pd
```

```python
df=pd.DataFrame({
    'city': ['New York','London','Paris','Tokyo','New York','Paris'],
    'price': [200,150,300,250,180,320]
})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df[['price','city_encoded']]
```
