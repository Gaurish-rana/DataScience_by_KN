Python logging is a built-in mechanism used to record messages about a program’s execution, helping developers understand what the program is doing and diagnose problems. Instead of using print() statements, logging allows you to capture information such as normal operation details, warnings, errors, and critical failures in a structured and configurable way. You can control where logs go (console, file, etc.), how detailed they are, and their format, making logging suitable for both development and production environments. Overall, Python logging helps in debugging, monitoring application behavior, and maintaining reliable, professional-quality software.

Python Logging Levels:

DEBUG – Very detailed messages used for debugging.
Helps developers track step-by-step execution.

INFO – Confirms that things are working as expected.
Used for normal, successful events.

WARNING – Indicates a potential problem or unexpected situation.
Program still runs, but attention is needed.

ERROR – A serious issue occurred and some functionality failed.
Program continues, but results may be incorrect.

CRITICAL – A very severe error that may stop the program.
Immediate action is required.

```python
import logging

# Configuring logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H-%M-%S',
    force=True
)

# log message with different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

```python
# Multiple Logging
import logging

# Create a logger for Module 1
logger1=logging.getLogger("module 1")
logger1.setLevel(logging.DEBUG)


# Create a logger for Module 2
logger2=logging.getLogger("module 2")
logger2.setLevel(logging.WARNING)

# Configure Logging Setting
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H-%M-%S',
    force=True
)
```

```python
# Log message with different loggers
logger1.debug("This is a debug message for module 1")
logger2.warning(" This is the warning message for module 2")
logger2.error("This is an error message")
```
