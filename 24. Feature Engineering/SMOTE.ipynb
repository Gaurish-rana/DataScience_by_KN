## SMOTE(Synthetic Minority Over-sampling Technique)

##### It is a method used in machine learning to solve the problem of imbalanced datasets, where one class has much fewer data points than the other. Instead of simply copying existing minority-class data (which can cause overfitting), SMOTE creates new, artificial data points. It works by selecting a data point from the minority class, finding its nearest minority-class neighbors, and then generating a new point by placing it somewhere between the original data point and one of its neighbors. This way, the new data looks realistic and follows the same pattern as the original minority class. By increasing the number of minority-class samples, SMOTE helps machine learning models learn the decision boundary better and improves performance in tasks like fraud detection, disease prediction, and spam classification.

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000, n_redundant=0, n_features= 2, n_clusters_per_class=1, 
    weights=[0.9], random_state=12
    )
```

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# Transform the dataset
oversample=SMOTE()
x,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```
