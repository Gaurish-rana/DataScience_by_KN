**Pandas** is a powerful Python library used for **data manipulation and data analysis**. It provides easy-to-use data structures like **Series** and **DataFrame**, which allow data to be cleaned, filtered, transformed, and analyzed efficiently. Pandas makes working with structured data such as CSV files, Excel sheets, and databases simple and intuitive. It is widely used in data science, machine learning, and analytics because it helps handle missing data, perform statistical analysis, and visualize trends quickly.


A **Series** in **Pandas** is a **one-dimensional labeled data structure** used to store a sequence of values along with their corresponding labels, called an **index**. It is similar to a column in a table or an enhanced Python list, but with the added benefit of indexing, which makes data access and manipulation easier. A Series can hold different data types such as integers, floats, strings, or even Python objects, and it is commonly used for handling and analyzing single-column data in Pandas.


```python
#Series
import pandas as pd

data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
# Create a series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
series=pd.Series(data,index=index)
print(series)
```

A **DataFrame** in **Pandas** is a **two-dimensional, labeled data structure** used to store and manage data in a **table-like format**, similar to an Excel sheet or a SQL table. It consists of **rows and columns**, where each column can hold different data types, and both rows and columns have labels (index and column names). DataFrames make it easy to organize, filter, analyze, and manipulate large datasets efficiently, which is why they are widely used in data analysis and data science.


```python
# Dataframe
# Create a Dataframe from a dictionary of list

data={
    'Name':['Dev','Rey','John'],
    'Age':[25,30,35],
    'City':['Bangalore','New York','Madrid']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# Create a dataframe from a list of dictionaries

data=[
    {'Name':'Jai','Age':26,'City':'Delhi'},
    {'Name':'Dev','Age':28,'City':'Manali'},
    {'Name':'Tej','Age':25,'City':'Jaipur'},
    {'Name':'Jack','Age':29,'City':'Daman'},
]
df=pd.DataFrame(data)
print(df)
```

The difference between `loc` and `iloc` in Pandas lies in how they access data. The `loc` method is **label-based**, meaning it selects rows and columns using their names or labels and includes the ending index in slicing. It also supports boolean conditions for filtering data. In contrast, `iloc` is **integer position–based**, meaning it selects data using numerical positions starting from zero, excludes the ending index in slicing, and does not use row or column labels.


```python
df.loc[0]
```

```python
df.iloc[0]
```

```python
# Accessing a specific element
df.at[1,'Transport_Type']
```

```python
# Accessing a specified element using iat
df.iat[2,2]
```

*Data manipulation and analysis* with Pandas involves cleaning, organizing, transforming, and examining data efficiently using Pandas’ powerful data structures like Series and DataFrames. With Pandas, you can easily load data from files such as CSV or Excel, handle missing values, filter rows and columns, sort and group data, create new columns, and perform statistical operations. It allows quick insights through functions like aggregation, merging, and reshaping datasets, making it an essential tool for data analysis, data science, and machine learning workflows.

```python
import pandas as pd

df=pd.read_csv('Product_Data_Pandas_Practice.csv')
df
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
# Hnadling Misiing values
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
# Filling missing values with the mean of the column
df['Rating_fill']=df['Rating'].fillna(df['Rating'].mean())
df
```

```python
df.dtypes
```

```python
# Renaming Columns  
df=df.rename(columns={'Product_ID':'PID'})
df
```

```python
# Change Datatypes
df['QuantitySold_new']=df['Quantity_Sold'].fillna(df['Quantity_Sold'].mean()).astype(int)# fillna is done as it has missing values
df
```

```python
# Apply Function
df['New_Rating']=df['Rating'].fillna(df['Rating'].mean()).apply(lambda x:x**2)
df
```

```python
# Data Aggregating and Grouping
Grouped_mean=df.groupby('Product_Name')['Price'].mean()
print(Grouped_mean)
```

```python
Grouped_sum=df.groupby(['Category','Product_Name'])['Price'].sum()
print(Grouped_sum)
```

```python
# Aggregating Multiple functions
Grouped_agg=df.groupby('Category')['Price'].agg(['mean','sum','count'])
print(Grouped_agg)
```

```python
# Merging and Joining Dataframes
# Create Sample Dataframes
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'],'Value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# Merge Dataframe on the 'Key Columns'
pd.merge(df1,df2,on='Key',how="inner")
```

```python
# Outer Join
pd.merge(df1,df2,on='Key',how="outer")
```

```python
# Left Join
pd.merge(df1,df2,on='Key',how="left")
```

```python
# Right Join
pd.merge(df1,df2,on='Key',how="right")
```

A **pickle file** is a way for Python to save data so it can be used again later without having to recreate or recompute it. It stores Python objects such as lists, dictionaries, DataFrames, or even machine learning models in a special binary format. The main purpose of a pickle file is to make programs faster and more efficient by allowing you to save the final result of your work and load it back exactly as it was. For example, instead of cleaning a dataset or training a model every time you run your code, you can pickle it once and reload it instantly whenever needed.
