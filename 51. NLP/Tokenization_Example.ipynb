**NLTK (Natural Language Toolkit)** is a widely used Python library for **Natural Language Processing (NLP)**. It provides a collection of tools, algorithms, and datasets that help computers understand, process, and analyze human language. NLTK is commonly used for tasks such as tokenization, stop-word removal, stemming, lemmatization, part-of-speech tagging, text classification, and sentiment analysis. Due to its simplicity and extensive documentation, it is widely used by students, researchers, and developers for learning and building NLP applications.


```python
import nltk
print(nltk.__version__)
```

```python
corpus="""Hello Welcome,to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
##  Tokenization
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize

documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization 
## Paragraph-->words
## sentence--->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

`word_tokenize()` and `wordpunct_tokenize()` are both tokenization methods available in NLTK, but they differ in how they split text. `word_tokenize()` is a more advanced and language-aware tokenizer that uses the Punkt sentence tokenizer and Treebank word tokenizer rules, making it suitable for most NLP tasks. It can correctly handle contractions such as "don't" by splitting it into `Do` and `n't`. In contrast, `wordpunct_tokenize()` is a simpler tokenizer based on regular expressions that separates words and punctuation marks purely by pattern matching. As a result, it splits "don't" into `Don`, `'`, and `t`. Therefore, `word_tokenize()` is generally preferred for accurate natural language processing, while `wordpunct_tokenize()` is useful for quick and simple text splitting without requiring additional language models.


```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

TreebankWordTokenizer is a tokenizer in NLTK that follows the tokenization rules of the Penn Treebank corpus. It is designed to split text into words while handling punctuation, contractions, and special cases more intelligently than simple tokenizers. For example, it separates contractions such as "don't" into do and n't, and treats punctuation marks as separate tokens when appropriate. Unlike wordpunct_tokenize(), which relies on simple regular expressions, TreebankWordTokenizer uses linguistically informed rules, making it more suitable for natural language processing tasks. In fact, word_tokenize() internally uses the Treebank tokenizer after sentence tokenization, which is why their outputs are often very similar.

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python

tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
